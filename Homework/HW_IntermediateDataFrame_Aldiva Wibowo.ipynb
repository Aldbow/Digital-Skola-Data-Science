```python
import pandas as pd
```

```python
# Membaca dataframe csv
df = pd.read_csv('https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv')
df.head()
```

```python
# manipulasi tabel dengan (filter dan sort values) dan mengurutkan dari umur tua ke muda juga tiket mahal ke murah
df.filter(items=['Name', 'Sex', 'Age', 'Ticket', 'Fare']).sort_values(by=['Age', 'Fare'], ascending=False)
```
