# Import Library

```python
import pandas as pd
```

# Create Dataframe

```python
# method 1 : list of lists
data_1 = [ ['Adi', 18, 'Apple'], ['Budi', 19, 'Mango'], ['Charlie', 22, 'Cherry'], ['Dino', 25, 'Durian'] ]

df_1 = pd.DataFrame(data_1, columns=['Name', 'Age', 'Fav Fruit'])
df_1
```

```python
# coba ganti index
df_1.index += 1

df_1
```

```python
# method 2 : dict of lists
data_2 = {'Name':['Adi','Budi','Candra','Dewi'],
          'Age':[10,20,35,22],
          'Fav Fruit':['Apple','Orange','Cherry','Avocado']}

df_2 = pd.DataFrame(data_2) # tidak perlu kasih column name karena sudah ambil dari key yang ada di dalam dict
df_2
```

```python
# method 3 : list of dicts
data_3 = [ {'Name':'Roman','Age':18,'Sport':'Soccer'},
          {'Name':'Ramon','Age':17,'Sport':'Basketball'},
          {'Name':'Rimin','Age':15,'Sport':'Ballet'} ]

df_3 = pd.DataFrame(data_3)
df_3
```

```python
# method 4 : Lists
list_1 = ['Adi','Budi','Candra','Dewi']
list_2 = [10,20,35,22]
list_3 = ['Apple','Orange','Cherry','Avocado']

data_4 = list(zip(list_1,list_2,list_3)) # menjadi list berisi beberapa tuple

df_4 = pd.DataFrame(data_4)
df_4.columns = ['Name','Age','Fruit'] # mengganti nama kolom dataframe
df_4.columns = ['Nama','Usia','Buah']
df_4
```
